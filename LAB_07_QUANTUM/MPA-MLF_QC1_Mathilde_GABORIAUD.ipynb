```python
# Create quantum circuit with 2 qubits and 2 classical bits
qc = QuantumCircuit(4, 2)
qc.x(0)
qc.x(1) 
qc.cx(1,2)
qc.cx(0,2)  # CNOT controlled by qubit 0 and targeting qubit 1
qc.ccx(0,1,3)
qc.measure(2,0)
qc.measure(3,1)
display(qc.draw())     # display a drawing of the circuit
job = sim.run(qc)      # run the experiment
result = job.result()  # get the results
# interpret the results as a "counts" dictionary
result.get_counts()
```

```python
from qiskit.visualization import plot_histogram
plot_histogram(result.get_counts())
```

```python
qc2 = QuantumCircuit(2, 2)
qc2.h(0) #variable qbit
qc2.measure(0,0)
qc2.measure(1,1)
display(qc2.draw())     # display a drawing of the circuit
job2 = sim.run(qc2)      # run the experiment
result2 = job2.result()  # get the results
# interpret the results as a "counts" dictionary
result2.get_counts()
```

```python
backend = Aer.get_backend('statevector_simulator')
qc2.x(1)
job3= backend.run(qc2)
result3=job3.result()
outputstate = result3.get_statevector(qc2,decimals=3)
print(outputstate)
```
